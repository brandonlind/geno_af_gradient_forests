# get performance data

[top](#home)

```python
# get directories
pooled_dirs = rt.dirs['pooled']

pooled_dirs
```

```python
# get the performance output files
res_files = []
for rep, repdirs in pooled_dirs.items():
    for repdir in repdirs:
        num_loci = repdir.split("/")[6]
        
        resdir = f'{repdir}/gradient_forests/validation/corrs'
        
        if not op.exists(resdir):
            print(ColorText(f'{rep} {num_loci}').fail().bold())
            continue
            
        files = fs(resdir, endswith='.pkl')
        res_files.extend(files)
        print(rep, num_loci, len(files))
```

<a id='comb'></a>
### combine performance data

[top](#home)

```python
def process_performance(pkl, source='pooled'):
    """Read in and add metadata to performance data."""
    import runtime_API as rt
    
    rt.params = params
    
    return rt.process_performance_pkl(pkl, source=source)

params = mvp.read_params_file()
dview['params'] = params
```

```python
for pkl in res_files:
    x = process_performance(pkl)
    break
    
x
```

```python
jobs = []
for pkl in res_files:
    jobs.append(
        lview.apply_async(
            process_performance, pkl
        )
    )
    
watch_async(jobs)
```

```python
all_data = pd.concat([j.r for j in jobs])

all_data
```

```python
# exclude 1-trait 1-nuisance jobs
data = all_data[all_data['plevel'] != '1-trait']

data
```
